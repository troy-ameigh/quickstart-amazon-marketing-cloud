### Create a workflow

```python
customerId = ''
workflowId = ''
query = """     

"""
workflowDetails = {
    "metadata": {
        "workflowName": '',
        "description": '' ,
    },
    "sqlQuery": query,
    "defaultPayload": {
            "timeWindowEnd": '' ,
            "timeWindowStart": '' ,
            "timeWindowType": '' ,
            "workflowExecutedDate": ''
    }
}
```

```python
workflow_interface.set_workflow_record(
    customerId=customerId,
    workflowId=workflowId,
    workflowDetails=workflowDetails
)
```
